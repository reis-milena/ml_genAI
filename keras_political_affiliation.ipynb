# Predict political parties with Keras

```python
import pandas as pd
```

# Treating data

```python
feature_names = ['party','handicapped-infants','water-project-cost-sharing',
                  'adoption-of-the-budget-resolution','physician-fee-freeze',
                  'el-salvador-aid','religious-groups-in-schools',
                  'anti-satellite-test-ban','aid-to-nicaraguan-contras',
                  'mx-missle','immigration','synfuels-corporation-cutback',
                  'education-spending','superfund-right-to-sue','crime',
                  'duty-free-exports','export-administration-act-south-africa']

voting_data = pd.read_csv('house-votes-84.data.txt',na_values = ['?'],
                          names = feature_names)
voting_data.head(10)
```

```python
voting_data.shape
```

```python
voting_data.dtypes
```

```python
voting_data.describe()
```

```python
#drop missing but attention to no bias data
voting_data.dropna(inplace=True)
voting_data.shape
```

```python
voting_data.describe()
```

```python
voting_data.replace(('y','n'), (1,0),inplace = True)
voting_data.replace(('democrat','republican'), (1,0),inplace = True)
```

```python
voting_data.head()
```

```python
all_features = voting_data.iloc[:,1:].values
all_classes = voting_data['party'].values
```

```python
all_features
```

```python
len(feature_names)
```

# Model

Keras + Scikit learn

```python
from sklearn.model_selection import cross_val_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from scikeras.wrappers import KerasClassifier
```

```python
all_features, all_classes
```

```python
def create_model():
    model = Sequential()
    model.add(Dense(16, activation='relu', input_shape = (16,))) #len(feature_names) -1
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss= 'binary_crossentropy', optimizer='rmsprop', metrics=['accuracy'])

    return model
```

```python
estimator = KerasClassifier(model= create_model, epochs = 100, verbose = 0)
```

```python
cv_scores = cross_val_score(estimator,all_features,all_classes,cv=10)
print(cv_scores.mean())
```

```python
def create_model2():
    model = Sequential()
    model.add(Dense(32, activation='relu', input_dim=16))
    model.add(Dense(16, activation='relu')) 
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss= 'binary_crossentropy', optimizer='adam', metrics=['accuracy'])

    return model
```

```python
estimator2 = KerasClassifier(model= create_model2, epochs = 100, verbose = 0)
cv_scores = cross_val_score(estimator2,all_features,all_classes,cv=10)
print(cv_scores.mean())
```
